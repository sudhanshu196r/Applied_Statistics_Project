```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Superstore Marketing Data.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors = 'coerce')
```

```python
median_date = df['Dt_Customer'].median()
```

```python
df['Dt_Customer'].fillna(median_date, inplace=True)
```

```python
df['Dt_Customer']
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Income']=df['Income'].astype(float)
```

```python
df['Income']
```

```python
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%m/%d/%Y')
```

Analyze the distribution of customer demographics (age, education, marital status) using descriptive statistics and visualizations.

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.histplot(2025-df['Year_Birth'], kde=True, bins=30, color='r')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.subplot(2,2,2)
sns.barplot(df['Education'].value_counts(), color='g')
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Education Level')

plt.subplot(2,2,3)
sns.barplot(df['Marital_Status'].value_counts(), color ='b')
plt.xlabel('Marital Status')
plt.ylabel('Count' )
plt.title('Marital Status Distribution')

plt.show()
```

# Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.:

```python
des_stat = df.describe()
print(pd.DataFrame(des_stat))
```

```python
#Total spendind
df['Total_Spending'] = df['MntFishProducts']+df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntGoldProds']+df['MntSweetProducts']
print(df['Total_Spending'])
```

```python
# Key Variable
Key_Variable=df[['Year_Birth','Income','Total_Spending']]
Key_Variable
```

```python
# Descriptive Statistics For Key Variable
stat=Key_Variable.describe()
print(pd.DataFrame(stat))
print('-----------------------------------------------------------------')
# Mode of key Variable
mode=Key_Variable.mode()
print('Mode of Key Variable:','\n',pd.DataFrame(mode))
print('----------------------------------------------------------------')
# Variance Of Key Variables
variance=Key_Variable.var()
print('Variance of Key Variable:','\n',pd.DataFrame(variance))
print('----------------------------------------------------------------')
```
